```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_final=pd.read_csv('https://raw.githubusercontent.com/JGuarnizo/Sales-Predictions/main/sales_predictions_clean.csv')
df_final.info()
```

```python
#Ventas por tienda
sales_outlet=df_final.groupby(['Outlet_Identifier'])[['Item_Outlet_Sales']].sum()
sales_outlet.plot.bar();
```

```python
#Ventas por categoría
sales_item=df_final.groupby(['Item_Type'])[['Item_Outlet_Sales']].sum()
sales_item.plot.bar();
```
